# Getting Data From CrisisNET

### Import Required Modules

```python
import requests
import pandas as pd

pd.set_option('display.max_columns', 50)
```

### Create a variable with your CrisisNET token (CHANGE THIS)

```python
token = '532d8dc4ed3329652f114b73'
```

### Setup the request headers

```python
headers = {'Authorization': 'Bearer ' + token}
```

### Setup the request URL

```python
url = 'http://devapi.crisis.net/item?%(tags)s&%(before)s&%(after)s&%(text)s&%(location)s&%(radius)s&%(limit)s&%(sources)s&%(licenses)s'
```

### Create a list of filters (CHANGE THIS)

```python
filters =   {'tags' : 'tags=', # a list of keywords seperated by comma
             'after' : 'after=', # a timestamp in ISO 8601 format
             'before' : 'before=', # a timestamp in ISO 8601 format
             'text' : 'text=', # words seperated by a '+'
             'location' : 'location=', # latitude and longitude
             'radius' : 'radius=', # a number in meters
             'limit' : 'limit=', 
             'sources' : 'sources=',
             'licenses' : 'licenses='
            }
```

### Create the formatted pull URL

```python
formattedURL = url % filters
```

### Check the formatted pull URL

```python
formattedURL
```

### Pull the data from CrisisNET

```python
r = requests.get(url, headers=headers)
```

```python
r
```

```python
pull_df = pd.DataFrame(r.json())
```

```python
pull_df
```

```python
data_df = pull_df['data'].apply(pd.Series)
data_df
```

```python
geo_df = data_df['geo'].apply(pd.Series)
geo_df
```

```python
geo_admin_df = geo_df['addressComponents'].apply(pd.Series)
geo_admin_df
```

```python
full_df = pd.concat([data_df[:], geo_admin_df[:], geo_df[:]], axis=1)
```

```python
full_df['latitude'] = full_df['coords'].str[1]
```

```python
full_df['longitude'] = full_df['coords'].str[0]
```
