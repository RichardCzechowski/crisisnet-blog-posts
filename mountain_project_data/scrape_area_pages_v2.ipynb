```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
import numpy as np
import os

cur_dir = os.path.dirname(os.path.realpath('__file__'))
```

```python
df_links = pd.read_csv(cur_dir + '/data/area_list.csv')
```

```python
title = []
geo = []
state = []
city = []
area = []
crag = []
wall = []
comments_name = []
comments_text = []
comments_loc = []
comments_date = []
```

```python
for link in df_links['area_url']:
    
    r = requests.get(link)
    soup = BeautifulSoup(r.content)
    
    names = []
    text = []
    loc = []
    date = []
        
    try:
        title.append(soup.find('h1').string)
        
        breadcrumb = soup.findAll('span',{'itemprop':'title'})
        if len(breadcrumb) == 5:
            state.append(breadcrumb[0].text)
            city.append(breadcrumb[1].text)
            area.append(breadcrumb[2].text)
            crag.append(breadcrumb[3].text)
            wall.append(breadcrumb[4].text)
        elif len(breadcrumb) == 4:
            state.append(breadcrumb[0].text)
            city.append(breadcrumb[1].text)
            area.append(breadcrumb[2].text)
            crag.append(breadcrumb[3].text)
            wall.append(np.NaN)
        elif len(breadcrumb) == 3:
            state.append(breadcrumb[0].text)
            city.append(breadcrumb[1].text)
            area.append(breadcrumb[2].text)
            crag.append(np.NaN)
            wall.append(np.NaN)            
        elif len(breadcrumb) == 2:
            state.append(breadcrumb[0].text)
            city.append(breadcrumb[1].text)
            area.append(np.NaN)
            crag.append(np.NaN)
            wall.append(np.NaN)
        else:
            state.append(np.NaN)
            city.append(np.NaN)
            area.append(np.NaN)
            crag.append(np.NaN)
            wall.append(np.NaN)
            
        geo_cell = soup.find(text='View Map')
        try:
            if geo_cell is not None:
                geo.append(geo_cell.findParent().findParent().contents[0].string[:-1])
            else:
                geo.append(np.NaN)
        except:
            geo.append(np.NaN)
        
        i = 0
        if len(soup.findAll('table', id=re.compile('^cm_'))) > 0:
            for comment in soup.findAll('table', id=re.compile('^cm_')):
                columns = comment.find_all('td')
                names.append(columns[0].find('a').text) 
                text.append(columns[1].text)
                if len(columns[0].findAll('br')) == 3:
                    loc.append(columns[0].find('br').nextSibling.text) 
                    date.append(columns[0].find('br').nextSibling.nextSibling.nextSibling.text) 
                    continue
                elif len(columns[0].findAll('br')) == 2:
                    loc.append(np.NaN)
                    date.append(columns[0].find('br').nextSibling.text)
                    continue
                i=i+1
                continue
        else:
            comments_name.append(np.NaN) 
            comments_text.append(np.NaN)
            comments_date.append(np.NaN) 
            comments_loc.append(np.NaN)
            i=i+1
            continue
                      
        comments_name.append(str(names))
        comments_text.append(str(text)) 
        comments_date.append(str(date))
        comments_loc.append(str(loc)) 

    except:
        comments_name.append(str(names))
        comments_text.append(str(text)) 
        comments_date.append(str(date))
        comments_loc.append(str(loc)) 
        continue
```

```python
print(len(title))
print(len(geo))
print(len(state))
print(len(city))
print(len(area))
print(len(crag))
print(len(wall))
print(len(comments_name))
print(len(comments_text))
print(len(comments_loc))
print(len(comments_date))
```

```python
df = pd.DataFrame()

df['title'] = title
df['url'] = df_links['area_url']
df['geo'] = geo
df['state'] = state
df['city'] = city
df['area'] = area
df['crag'] = crag
df['wall'] = wall
df['comments_name'] = comments_name
df['comments_text'] = comments_text
df['comments_date'] = comments_date
df['comments_loc'] = comments_loc
```

```python
def lat_long_split(geo_df):

    lat = []
    lon = []

    for row in geo:
        try:
            lat.append(row.split(',')[0])
            lon.append(row.split(',')[1])
        except:
            lat.append(np.NaN)
            lon.append(np.NaN)

    geo_df['latitude'] = lat
    geo_df['longitude'] = lon
```

```python
lat_long_split(df['geo'])
```

```python
df.to_csv(cur_dir + '/data/area_data_2.csv', index=False)
```
